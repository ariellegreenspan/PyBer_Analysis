```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a scatterplot of the data using the plot function
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create the a scatterplot of the data using the scatter function
plt.scatter(x_axis, y_axis)
```

```python
# Change the color of the markers to red. Add legend saying "Chicago." Switch the axis so that the Fare($) data is on the x-axis.
plt.plot(y_axis, x_axis, 'o', color = "red", label = "Chicago")
# Add a title and axes labels.
plt.title("Pyber Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Month")
# Add a limit for the x-axis data.
plt.xlim(0, 50)
# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
# Add the legend.
plt.legend()
```
